```python
from scipy import stats
```

```python
#continuous vs continuous (two sample)
def continuous_vs_continuous(col1,col2):
  t=0
  f=0
  for i in range(31):
    samp1=bodyfat[col1].sample(frac=0.03)
    samp2=bodyfat[col2].sample(frac=0.03)
    ttest,pvalue=stats.ttest_ind(samp1,samp2)
    if pvalue<0.02:
      f=f+1
    else:
      t=t+1

  if t>f:
    return True
  else:
    return False
```

```python
#continuous(col1) vs category(col2)
def continuous_vs_categories(col1,col2):
  group=bodyfat[col2].unique()
  data={}
  for i in group:
    data[i]=bodyfat[col1][bodyfat[col2]==i]
  fvalue,pvalue=stats.f_oneway(*[i for i in data.values()])
  return False if pvalue<0.2 else True
```

```python
  final={}
  for i in bodyfat.columns:
    final[i]={}
    for j in bodyfat.columns:
      if i == j:
              final[i][j] = 0  # No self-comparison
              continue
      if (i in continuous) and (j in continuous):
        result=continuous_vs_continuous(i,j)
      elif (i in continuous) and (j in categories):
        result=continuous_vs_categories(i,j)
      elif (i in categories) and (j in continuous):
        result=continuous_vs_categories(j,i)
      else:
              result = False  # Catch all unmatched types

      final[i][j] = int(result)
```
